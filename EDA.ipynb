## 1.2. Data Cleaning

```python
# Identifying missing values...

# Below, I replaced missing values with NaN to work with NaN functions
df = df.replace("", np.nan) 
df = df.replace("Hidden", np.nan) 

null_values_df = df.isnull()

print("Null Values in Each Column:\n")
print(null_values_df.sum())
```

```python
# Now, assigning data types to columns...

numeric_columns = ["Purchase_Amount", "Average_Spending_Per_Purchase",
                   "Purchase_Frequency_Per_Month", "Brand_Affinity_Score"]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

#df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"].dt.date, errors='coerce')

df['Year'] = df['Year'].astype('Int64')

df.info()
```

```python
# Filling NaN values...

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'] = df['Age'].round().astype(int)

fill1 = ['Purchase_Amount', 'Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score', 'Year']
for column in fill1:
    df[column] = pd.to_numeric(df[column], errors='coerce')
    df[column].fillna(df[column].mean(), inplace=True)
    df[column] = df[column].round().astype(int)

fill2 = ['Customer_ID', 'Gender', 'Income_Level','Address', 'Transaction_ID', 'Purchase_Date',
                   'Product_ID', 'Product_Category', 'Brand', 'Product_Category_Preferences', 'Month', 'Season'] 
df[fill2] = df[fill2].fillna('Unknown').astype(str)

df.info()
```

```python
# Cleaning Address field from "\n" characters...

df['Address'] = df['Address'].str.replace('\n', ' ')
print(df['Address'])
```

## 1.3. Data Transformation

```python
# Introducing new columns: Feature 1...

df['Purchase_Income_Ratio'] = df['Purchase_Amount'] / df.groupby('Income_Level')['Purchase_Amount'].transform('mean')
df[['Income_Level', 'Purchase_Amount', 'Purchase_Income_Ratio']]
```

```python
# Introducing new columns: Feature 2...

df['Opposite_Gender_Brand_Purchase_Count'] = ''

brand_gender_counts = df[df['Brand'] != 'Unknown'].groupby(['Brand', 'Gender']).size().reset_index(name='Count')

for index, row in df.iterrows():
    brand = row['Brand']
    gender = row['Gender']

    opposite_gender = 'Male' if gender == 'Female' else 'Female' if gender == 'Male' else 'Male'

    count_opposite_gender = brand_gender_counts[
        (brand_gender_counts['Brand'] == brand) & (brand_gender_counts['Gender'] == opposite_gender)]['Count'].values

    df.at[index, 'Opposite_Gender_Brand_Purchase_Count'] = count_opposite_gender[0] if count_opposite_gender.size > 0 else 0

df[['Brand', 'Gender', 'Opposite_Gender_Brand_Purchase_Count']]
```

```python
# Introducing new columns: Feature 3...

df['Customer_Age_Group'] = ''

c1 = (df['Age'] >= 0) & (df['Age'] < 18)
c2 = (df['Age'] >= 18) & (df['Age'] < 25)
c3 = (df['Age'] >= 25) & (df['Age'] < 60)
c4 = (df['Age'] >= 60) & (df['Age'] < 75)
c5 = (df['Age'] >= 75)

df.loc[c1, 'Customer_Age_Group'] = 'Child'
df.loc[c2, 'Customer_Age_Group'] = 'Teenager'
df.loc[c3, 'Customer_Age_Group'] = 'Adult'
df.loc[c4, 'Customer_Age_Group'] = 'Senior'
df.loc[c5, 'Customer_Age_Group'] = 'Elderly'

df[['Age', 'Customer_Age_Group']]
```

## 3.1. K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#1.Define the number of clusters (k):
#Analyze the elbow plot to determine the optimal number of clusters based on the sum of squared distances within each cluster.
#Consider silhouette analysis to evaluate the quality of clusters formed at different k values.
#2.Apply K-Means algorithm:
#Implement K-Means with the chosen k value to segment customers into distinct clusters based on their purchase behavior and preferences.
#3.Analyze cluster characteristics:
#Investigate the key features of each cluster, such as average purchase amount, brand affinity, and product category preferences.
#Identify significant differences and similarities between the clusters.
# making a float column for Product Category Preference column
map_values = {'Low': 0, 'Medium': 0.5, 'High': 1}
df['Product_Category_Preferences_Value'] = df['Product_Category_Preferences'].map(map_values)
df['Product_Category_Preferences_Value'].fillna(df['Product_Category_Preferences_Value'].mean(), inplace=True)

# K-Means Clustering

# Defining the columns for K-means clustering

cols = ['Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score', 'Product_Category_Preferences_Value']
cluster_cols = df[cols]

# Standardizing the columns

scaler = StandardScaler()
standardized_cluster_cols = scaler.fit_transform(cluster_cols)

# Analyzing the elbow plot

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(standardized_cluster_cols)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Plot for K-Means')
plt.show()

# Silhouette analysis to determine quality of clusters k

silhouette = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(standardized_cluster_cols)
    silhouette.append(silhouette_score(standardized_cluster_cols, kmeans.labels_))

plt.plot(range(2, 11), silhouette, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for K-Means')
plt.show()

# On the basis of elbow plot and silhouette analysis choosing optimal number of clusters k
clusters = 2

# kmeans with k clusters
kmeans = KMeans(n_clusters=clusters, random_state=42)
kmeans.fit(standardized_cluster_cols)

# adding kmeans column to database
df['KMeans_Cluster'] = kmeans.labels_

# Displaying the cluster characteristics
cluster_characteristics = df.groupby('KMeans_Cluster')[cols].mean()
print(cluster_characteristics)
```

## 3.2. DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
# DBSCAN CLustering
#1.Define eps and MinPts parameters:
#Experiment with different values of eps(neighborhood radius)and MinPts(minimum number of points within eps)to identify the best configuration for clustering.
#Utilize silhouette analysis or other cluster quality metrics to evaluate different parameter combinations.
#2.Apply DBSCAN algorithm:
#Implement DBSCAN with the chosen parameters to discover clusters of customers based on their density and spatial distribution.
#3.Analyze cluster characteristics:
#Explore the characteristics of each cluster, including size, density, and distribution within the data space.
#Compare the clusters formed by DBSCAN with those identified by KMeans.
# Visualizing the k-distance graph to determine optimal eps
def plot_k_distance(data, k_range=range(1, 11)):
    nn = NearestNeighbors(n_neighbors=max(k_range)).fit(data)
    distances, _ = nn.kneighbors(data)

    plt.figure(figsize=(10, 6))
    for k in k_range:
        plt.plot(sorted(distances[:, k - 1]), label=f'k={k}')

    plt.xlabel('Data Point Index')
    plt.ylabel('Distance to k-th Nearest Neighbor')
    plt.title('K-Distance Graph')
    plt.legend()
    plt.show()

# Plotting the k-distance graph
plot_k_distance(standardized_cluster_cols)

# Experimenting with different values of eps and MinPts, and choose optimal parameters
best_silhouette_score = -1
optimal_eps = 0
optimal_min_samples = 0

for eps in [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]:
    for min_samples in [1, 3, 5, 7, 9]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan_labels = dbscan.fit_predict(standardized_cluster_cols)

        if len(set(dbscan_labels)) > 1:
            silhouette_avg = silhouette_score(standardized_cluster_cols, dbscan_labels)
            print(f"eps={eps}, min_samples={min_samples}, Silhouette Score: {silhouette_avg}")

            if silhouette_avg > best_silhouette_score:
                best_silhouette_score = silhouette_avg
                optimal_eps = eps
                optimal_min_samples = min_samples

print(f"\nOptimal Parameters - eps: {optimal_eps}, min_samples: {optimal_min_samples}, best_silhouette_score: {best_silhouette_score}")

# Apply DBSCAN with the chosen parameters
dbscan = DBSCAN(eps=optimal_eps, min_samples=optimal_min_samples)
dbscan_labels = dbscan.fit_predict(standardized_cluster_cols)

# Add DBSCAN cluster labels to the original dataset
df['DBSCAN_Cluster'] = dbscan_labels

# Visualize the clusters in 2D (assuming X_scaled has 2 features)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=standardized_cluster_cols[:, 0], y=standardized_cluster_cols[:, 1], hue=dbscan_labels, palette='viridis', s=50)
plt.title('DBSCAN Clusters')
# plt.xlabel('Column 1')
# plt.ylabel('Column 2')
plt.show()

# Display the cluster characteristics
dbscan_cluster_characteristics = df.groupby('DBSCAN_Cluster')[cols].mean()
print(dbscan_cluster_characteristics)
```

## 3.3. K-Means++ Clustering

```python
#1.Apply K-Means++ algorithm:
#Implement K-Means++ with the desired number of clusters to initialize centroids strategically for improved clustering performance.
#2.Compare results to K-Means:
#Analyze the resulting clusters formed by K-Means++ and compare them to those generated by regular K-Means.
#Evaluate the effectiveness of K-Means++ in achieving better cluster quality and convergence speed.
# K-Means++ Clustering

kmeans_pp = KMeans(n_clusters=clusters, init='k-means++', random_state=42)
kmeans_pp.fit(standardized_cluster_cols)

# Add K-Means++ cluster labels to the original dataset
df['KMeans++_Cluster'] = kmeans_pp.labels_

# Display the cluster characteristics
kmeans_pp_cluster_characteristics = df.groupby('KMeans++_Cluster')[cols].mean()
print(kmeans_pp_cluster_characteristics)

# Comparing results between K-Means and K-Means++
compare_clusters = pd.DataFrame({
    'KMeans_Cluster': df['KMeans_Cluster'],
    'KMeans++_Cluster': df['KMeans++_Cluster']
})

print(compare_clusters)
#print(compare_clusters.head())
```
